```python
def hypothesis(x,theta):
    return theta[0] + theta[1]*x

def error(x,y,theta):
    err = 0
    rows = x.shape[0]
    
    for i in range(rows):
        err += (hypothesis(x[i],theta) - y[i])**2
    
    return err

def grad(x,y,theta):
    
    gradient=np.array([0.0,0.0])
    rows = x.shape[0]
    
    for i in range(rows):
        gradient[0] += (hypothesis(x[i],theta) - y[i])
        gradient[1] += (hypothesis(x[i],theta) - y[i])*x[i]
    
    return gradient
    
def gradientDescent(x,y):
    
    theta = np.array([2.0, 0.0])
    itr = 0
    while(itr < 100):
        theta = theta - 0.0001*grad(x,y,theta)
        itr += 1
        
    return theta
```

```python
def hypothesis(x,theta):
    return theta[0] + theta[1]*x


def error(X,Y,theta):
    
    m = X.shape[0]
    error = 0
    
    for i in range(m):
        hx = hypothesis(X[i],theta)
        error += (hx-Y[i])**2
        
    return error

def gradient(X,Y,theta):
    
    grad = np.zeros((2,))
    m = X.shape[0]

    for i in range(m):
        hx = hypothesis(X[i],theta)
        grad[0] +=  (hx-Y[i])
        grad[1] += (hx-Y[i])*X[i]
        
    return grad
    
#Algorithm
def gradientDescent(X,Y,learning_rate=0.0001):
    
    theta = np.array([-2.0,-5.0])
    
    itr = 0
    max_itr = 100
    
    error_list = []
    theta_list = []
    
    while(itr<=max_itr):
        grad = gradient(X,Y,theta)
        e = error(X,Y,theta)
        error_list.append(e)
        
        theta_list.append((theta[0],theta[1]))
        theta[0] = theta[0] - learning_rate*grad[0]
        theta[1] = theta[1] - learning_rate*grad[1]
        
        
        itr += 1
        
    
    return theta,error_list,theta_list

    
```

```python
newtheta,error_list,theta_list = gradientDescent(x,y)
```
