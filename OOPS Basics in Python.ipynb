```python
class Dog:
    """This is a Docstring in class"""
    ## Class Variables -> Common for all instances(object) of the class
    ## Public
    color = "brown"
    ## Private
    __age = 9
    
    ## Constructor
    def __init__(self,a,b):
        ## Private Instance variable
        self.__breed = a
        ## Public Instance variable
        self.name = b
        self.skills = []
    
    ## Public Method
    def addSkills(self,skill):
        """Docstring for function"""
        self.skills.append(skill)
    
    ## Private Method
    def __aboutDog(self):
        print("Dog name: ",self.name)
        print("Dog breed: ",self.__breed)
        print("Dog age: ",self.__age)
        print("Dog color: ",self.color)
        print("Dog skills: ",self.skills)
        
    ## Public Method
    def introduce(self):
        self.__aboutDog()

##############################################################
d1 = Dog("Golden Retriver","Fluffy")
d1.addSkills("High Jump")

d2 = Dog("German Shepard","Bruno")
d2.addSkills("Roll Over")

## d1.__aboutDog() -> Not accessible
d1.introduce()
print()
## d2.__aboutDog() -> Not accessible
d2.introduce()
print("#############################################################")
##############################################################
##############################################################
## Accessing The variables
print(d1.color)
## print(d1.__age) -> private
## print(d1.__breed) -> private
print(d1.skills)
print(d1.name)
print("#############################################################")
##############################################################
d1.color = "White"
d1.__age = 10 ## This does not work, but also does not give error

d2.name = "Husky"
d1.name = "Maguna"

d1.introduce()
print()
d2.introduce()
print("#############################################################")

d1.__class__.color = "White"
d1.__class__.__age = 10
d1.introduce()
print()
d2.introduce()
print("#############################################################")

d1.__class__.color = "White"
d1.__class__.__age = 10
d2.__class__.color = "Black"
d2.__class__.__age = 11
d1.introduce()
print()
d2.introduce()
print()
```
